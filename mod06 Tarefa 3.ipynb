#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
#  Criando a função que vai classficar
def classificador(nota: float) -> str:
    if 8<= nota <= 10:
        return 'normal'
    elif 6<= nota<=7:
        return 'asfixia leve'
    elif 4 <= nota <=5:
        return 'asfixia moderada'
    elif 0<= nota <=3:
        return 'asfixia grave'
```

```python
df['CLASS_APGAR5'] = df['APGAR5'].apply(classificador)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
freq_sexo = pd.crosstab(df['APGAR5'],df['SEXO'])
freq_sexo
```

```python
cross_tab = pd.crosstab(index=[df['GESTACAO']], columns=[df['APGAR5'], df['SEXO']], margins=True, margins_name='Total')
cross_tab
```

```python
cross_tab.sum(axis=1)
```

```python
cross_tab.plot.bar()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Tabela Cruzada
freq = pd.crosstab(df['APGAR5'],df['GESTACAO'] )
```

```python
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq.sum(axis=1)
```

```python
freq.pctlin = freq.div(freq.sum(axis=1), axis=0)
freq.pctlin
```

```python
freq.pctlin.sum(axis=1)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Plota o gráfico de dispersão
sns.regplot(x='APGAR5', y='PESO', data=df, scatter_kws={'s': 100}, ci=None, color='coral')

# Adiciona título e rótulos aos eixos
plt.title('Relação entre APGAR5 e PESO')
plt.xlabel('APGAR5')
plt.ylabel('PESO')

# Exibe o gráfico
plt.show()
```

Este código usa sns.regplot para criar um gráfico de dispersão com uma linha de regressão que mostra a relação entre as colunas APGAR5 e PESO. A linha de regressão ajuda a visualizar a tendência geral entre essas variáveis. Se a linha estiver inclinada para cima, indica uma correlação positiva, enquanto se estiver inclinada para baixo, indica uma correlação negativa. A ausência de uma inclinação clara pode sugerir uma fraca correlação. Como podemos observar, há sim essa correlação.
